**CÃ¢u 6:Táº¡o cÃ¡c biáº¿n TBM1, TBM2, TBM3 tÆ°Æ¡ng á»©ng vá»i trung bÃ¬nh mÃ´n cá»§a cÃ¡c nÄm lá»p 10, 11 vÃ  12. â¢ CÃ´ng thá»©c tÃ­nh: TBM = (T2 + L + H + S + V2 + X + D + N) / 10**

**CÃ¢u 7:Táº¡o cÃ¡c biáº¿n xáº¿p loáº¡i XL1, XL2 vÃ  XL3 dá»±a trÃªn TBM1, TBM2 vÃ  TBM3 cho tá»«ng nÄm lá»p 10, 11, 12 nhÆ° sau: â¢ Nhá» hÆ¡n 5.0 xáº¿p loáº¡i: yáº¿u (kÃ­ hiá»u lÃ  Y) â¢ Tá»« 5.0 Äáº¿n dÆ°á»i 6.5: trung bÃ¬nh (kÃ­ hiá»u lÃ  TB) â¢ Tá»« 6.5 Äáº¿n dÆ°á»i 8.0: khÃ¡ (kÃ­ hiá»u lÃ  K) â¢ Tá»« 8.0 Äáº¿n dÆ°á»i 9.0: giá»i (kÃ­ hiá»u lÃ  G) â¢ Tá»« 9.0 trá» lÃªn: xuáº¥t sáº¯c (kÃ­ hiá»u lÃ  XS)**

```python
for i in ['1','2','3']:
  Dataframe.loc[Dataframe['TBM'+i] < 5, 'XL'+i] = 'Y'
  Dataframe.loc[(Dataframe['TBM'+i] >= 5) & (Dataframe['TBM'+i] < 6.5) , 'XL'+i] = 'TB' 
  Dataframe.loc[(Dataframe['TBM'+i] >= 6.5) & (Dataframe['TBM'+i] < 8) , 'XL'+i] = 'K'
  Dataframe.loc[(Dataframe['TBM'+i] >= 8) & (Dataframe['TBM'+i] < 9) , 'XL'+i] = 'G'
  Dataframe.loc[(Dataframe['TBM'+i] >= 9), 'XL'+i] = 'XS'  
print(Dataframe)
```
